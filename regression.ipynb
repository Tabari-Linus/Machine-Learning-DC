```python
import pandas as pd
sales_df = pd.read_csv('supervise learning/datasets/advertising_and_sales_clean.csv')
sales_df.head()
```

```python
X = sales_df[['radio']].values
y = sales_df['sales'].values
```

```python
# x_radio = X[:,0]
# x_radio_shaped = x_radio.reshape(-1,1)
x_radio_shaped = X.reshape(-1,1)
x_radio_shaped.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(x_radio_shaped,y)
```

```python
prediction = reg.predict(x_radio_shaped)
prediction
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_radio_shaped,y)
plt.plot(x_radio_shaped, prediction, color="red")
plt.xlabel("Radio Expenses ($)")
plt.ylabel("Sales ($)")

plt.show()
```
